**Data Exploration**

```python
## Changing data types
data['id'] = data['id'].astype(object)
```

```python
data.info()
```

```python
data['id']
```

```python
## Identify outliers in price
sns.boxplot(data=data, x='price')
```

```python
## Boxplot graph with price < value_outlier
df = data[data['price']<1500]
sns.boxplot(data=df, x='price')
```

```python
## Creating a histogram on Distribution of price

plt.figure(figsize=(8,8))
sns.histplot(data=df, x='price', bins=100)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel("Frequency")
plt.show()
```

```python
df.columns
```

```python
## Grouping of data
df.groupby(by ='neighbourhood_group')['price'].mean()
```

```python
df.groupby(by ='neighbourhood_group')['price'].count()
```

**Feature Engineering**

```python
## Creating new column
df['price per beds'] = df['price']/df['beds']
```

```python
## Creating barplot of Price by Neighbourhood Group and Room Type

sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price per beds', hue='room_type')
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='room_type')
plt.title('Price and Reviews Dependency')
```

```python
## Heat Map :- Correlation of one variable to other variable for numerical columns

corr = df[['latitude','longitude', 'price','minimum_nights','number_of_reviews', 'reviews_per_month', 'availability_365']].corr()
corr
```
